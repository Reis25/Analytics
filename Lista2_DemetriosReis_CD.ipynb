9. Usando os dados a seguir e um classificador, a qual classe o texto “what do you
mean” deveria ser classificado? (1 ponto)

![texto alternativo](https://raw.githubusercontent.com/Reis25/Testes_py/master/Captura%20de%20tela%20de%202020-07-21%2018-43-39.png)

```
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB, BernoulliNB, MultinomialNB

from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score


data_frame = pd.read_csv("ciencia_dos_dados.csv")


```

```
data_frame.head(8)
```

```
data_frame['classe_texto'] = data_frame['classe_texto'].replace(['stmt', 'question'], [1, 0])

data_frame.head(8)
```

```
x = data_frame['text']

y = data_frame['classe_texto'] 

from sklearn.feature_extraction.text import TfidfVectorizer  

v = TfidfVectorizer(use_idf = True)  

x = v.fit_transform(x.astype('U')).toarray()  
```

```
num_train = 5

x_train = x[:num_train]
x_test = x[num_train:]

y_train = y[:num_train]
y_test = y[num_train:]

gnb = GaussianNB()

model = gnb.fit(x_train, y_train)

predicted = model.predict(x_test)

print('Predito:', predicted)
print('Real:', list(y_test))
```

```
print(accuracy_score(predicted, list(y_test)))
```

```
test = v.transform(['what do you mean']).toarray()

prediction = model.predict(test)

print("A classe que 'what do you mean' pertence é: " ,prediction)
```
